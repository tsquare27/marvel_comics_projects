```python
import pandas as pd
import requests as req
import hashlib
import time


my_private = "984f159687eb3ede36d3d88136521b55bff75476"
my_public =  "b3e5790f7464d6954dfd0398664b80f1"
# endpoint url 
BASEURL = "http://gateway.marvel.com"

characters =  "/v1/public/characters"
character_endpoint  = BASEURL+characters
comics ="/v1/public/comics"
comics_endpoint = BASEURL+comics

# This function to get the data in JSON from any given api point
def get_request(url):
    
    # Generate timestamp and encrypt the api keys using the md5 method
    ts = str(time.time()) # convert the timestamp to strings
    
    # Get the current time as a timestamp
    hash_input = ts + my_private + my_public 
    
    hash_result = hashlib.md5(hash_input.encode()).hexdigest()

    # Request headers (this depends on the API you're using)
    params = {   
                "ts":ts,    
                "apikey": my_public,    
                "hash": hash_result,
            }
    # Make the request
    response = req.get(url,params=params)

    # Check for successful response
    if response.status_code == 200:    
        # Convert the response to JSON (assuming the API returns JSON)
        data = response.json()   
        df= pd.DataFrame(data['data']['results'])
    # Show first few rows of the DataFrame
    
    else:
        print(f"Failed to retrieve data. Status code:{response.status_code}")
    return df
```

```python
characters_df =get_request(character_endpoint)
comics_df =get_request(comics_endpoint)


character_names = characters_df['name']  
expand_comics = pd.DataFrame(characters_df['comics'].tolist())
comic_count = expand_comics['available']

df = pd.DataFrame(
                    {'Character name':character_names, 'Comic Count':comic_count}
                )
df.sort_values('Comic Count',ascending=False).reset_index().drop(columns='index')

# Get the characters ID
# id = pd.DataFrame(data['id'])

#character_list = []

# #loop through each ID 
# for index, row in id.iterrows():
#     id_number = row['id']
#     base_url =f"{character_endpoint}/{id_number}/comics"
#     comics_baseurl.append(base_url)
    
#     # Make the GET request to the API
# #     resp = get_request(base_url)

# # res = pd.DataFrame(resp)
# # res
# print(comics_baseurl)
```
